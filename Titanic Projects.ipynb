```python
df.head() #in these Ticket having objective and int data types which is not important to predict the output so we drop that
df.drop(["Ticket"],inplace=True,axis=1)
```

```python
df.head()
```

```python
#now we areb going to performe labelEncoding on objective columnsn means Sex and Embarked
from sklearn.preprocessing import LabelEncoder
cols=["Sex","Embarked"]
#create object of LabelEncoder
le=LabelEncoder()
for col in cols:  #because cols stored SEx and Embarked
    df[col] = le.fit_transform(df[col])
```

```python
df.head()  #now Its converted
```

```python
#Select Input and output from dataset df
x=df.drop("Survived",axis=1) #input variable
y=df["Survived"] #output means target
```

```python
#train_test_split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
#for better performance i convert all the value in the same unit.....use standardscaler
from sklearn.preprocessing import StandardScaler
#create object of StandardScaler
ss=StandardScaler()
x_train=ss.fit_transform(x_train)    #training data
x_test=ss.transform(x_test)      #testing data
```

```python
x_train.shape  #70 % data to train
```

```python
x_test.shape  #30 % data to test
```

```python
#First we balanced the data 
#Sampling Technique
#no we see the our model is balanced or not 
df["Survived"].value_counts()
```

```python
#now  we use RandomOverSampling Technique for this dataset
#we call the RandomOverSampling from the imblearn package
from imblearn.over_sampling import RandomOverSampler
```

```python
#create object of random over sampler
ros = RandomOverSampler()
```

```python
#fit _resample() inbuilt method of RandomOversamle() class   balanced the training data
x_train,y_train=ros.fit_resample(x_train,y_train)
```

```python
#to balance test dataset
x_test,y_test = ros.fit_resample(x_test,y_test)
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

# LogisticRegression 

```python
#Create the baseline model: logisticRegression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
#create function 
def create_model(model):
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    print(classification_report(y_test,y_pred))
    print(confusion_matrix(y_test,y_pred))
    return model
```

```python
#Now we create the object of logisticRegression
lr=LogisticRegression()
```

```python
#Now we call the function
lr=create_model(lr)
```
